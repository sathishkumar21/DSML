```python
import pandas as pd
df_mcd = pd.read_excel('https://github.com/insaid2018/Term-1/raw/master/Data/Casestudy/mcdonalds_india_menu.xlsx')
df_mcd.head()
```

```python
df_mcd['Category'].value_counts().plot.bar(figsize=(15,7),colormap='Pastel2',fontsize=10,yticks=np.arange(0,19,2))
plt.xlabel("Category")
plt.ylabel("No. of Items")
plt.title("Bar Chart showing category")
plt.savefig("barchart.jpg")
```

```python
(df_mcd['Category'].value_counts()/len(df_mcd)).plot(kind='bar',figsize=(16,7),colormap='Wistia',fontsize=13)
#(df_mcd['Category'].value_counts() / len(df_mcd)).plot(kind='bar', figsize=(16,7), colormap='magma', fontsize=13)

plt.xlabel("Category")
plt.ylabel('Proportion of Items')
plt.title('Bar chart')
```

```python
df_mcd['Energy'].plot(kind='line',figsize=(15,7),color='blue', fontsize=14, linestyle='-')
plt.xlabel('Index')
plt.ylabel('Energy (kcal)')
plt.title('Line chart showing the variation of energy across different items')
```

```python
df_mcd['Serve_Size'].head(10)
```

```python
bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
df_mcd['Serve_Size'].plot(kind='hist', figsize=(15,7),bins=bins,yticks=np.arange(0,22,3),xticks=bins)

plt.xlabel("Serve")
plt.ylabel("Frequency")
plt.title("'Histogram showing the distribution of Serve Size values'")
```

```python
space=np.ones(11)/10
df_mcd['Category'].value_counts().plot(kind='pie',explode=space,figsize=(10,10),wedgeprops=dict(width=0.15),legend=True,startangle=160,shadow=True,autopct='%3.1f%%')

plt.ylabel('Category')
plt.title('Donut Plot showing the proportion of each Category value')
```

```python
df_mcd.plot(kind='scatter',x='Carbohydrates',y='Energy',figsize=(10,6))
plt.title('Scatter plot showing the variation of Energy with Carbohydrates present in each item')
```

```python
plt.figure(figsize=(12,7))
sns.regplot(data=df_mcd,x='Carbohydrates',y='Energy')
plt.title('Regplot showing the variation of Energy with Carbohydrates present in each item')
```

```python
df_mcd.groupby(['Category'])['Total_Fat'].plot(kind='bar',figsize=(15,7),color='orange')
df_mcd.groupby(['Category'])['Saturated_Fat'].plot(kind='bar',figsize=(15,7),color='red')
df_mcd.groupby(['Category'])['Trans_Fat'].plot(kind='bar',figsize=(15,7),color='green',)
                                               
plt.xlabel('Category')
plt.ylabel('Fat Content')
plt.title('Stacked Bar Chart showing the Fat content of each Category')
plt.legend(['Total_Fat','Saturated_Fat','Total_Fat'])
```

```python
#df_mcd.groupby(['Serve_Size'])['Cholestrol'].plot(kind='line',figsize(15,7),color('green'))
df_mcd.groupby(['Serve_Size'])['Cholestrol'].mean().plot(kind='line', figsize=(15, 7), color='green')
df_mcd.groupby(['Serve_Size'])['Total_Fat'].mean().plot(kind='line', figsize=(15, 7), color='red')
#df_mcd.groupby(['Serve_Size'])['Total_Fat'].plot(kind='line',figsize(15,7),color('green'))  

plt.ylabel("Value")
plt.title('Bivariate Linear Chart showing the variation of Cholestrol and Total_Fat with Serve_Size')
plt.legend(['Cholestrol (in mg)', 'Total_Fat (in g)'])
```

```python
# Using seaborn's pairplot function to plot the pairplot for specific columns of the dataset.

sns.pairplot(data=df_mcd[['Energy','Sugars','Carbohydrates','Protein']],size=2.5,diag_kind='kde')
```
